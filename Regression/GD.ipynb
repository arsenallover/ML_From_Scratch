```python
import numpy as np
```

```python
X = np.array((1,2,3,4,5))
X.shape
y = np.array((1,2,3,4,5))*10+ np.random.normal(1, 5, 5)*0.5
```

```python
import matplotlib.pyplot as plt
plt.plot(X, y)
```

```python
w = 1
b = 1
alpha = 0.1

for i in range(100):
    yhat = w*X+b
    error = 0.5*np.sum((y-yhat)**2)
    if i%10 ==0:
        print(b)
    dl_dw =  -2*alpha*(y-yhat)@X.T
    dl_dc = -2*alpha*np.sum(y-yhat, axis =0)
    w = w-dl_dw/5
    b = b- dl_dc/5
print(w,b)
```
