## Concept

* Density based outlier detection
* Compares local density of points with local density of its neighbours
    * when density of that point is lower -> outlier
* Points in low density regions have high LOF value -> outliers

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs
from numpy import quantile, where, random
import matplotlib.pyplot as plt

np.random.seed(42)

x, _ = make_blobs(n_samples=200, centers=1, cluster_std=.3, center_box=(10,10))
plt.scatter(x[:,0], x[:,1])
plt.show()

lof = LocalOutlierFactor(n_neighbors=20, contamination=.03)

y_pred = lof.fit_predict(x)


lofs_index = where(y_pred==-1)
values = x[lofs_index]

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0],values[:,1], color='r')
plt.show()
```
